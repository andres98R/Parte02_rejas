```python
def costoxpersona(tipo, distancia):
    if tipo == 1:
        return distancia*2
    elif tipo == 2:
        return distancia*2.5
    elif tipo == 3:
        return distancia*3

def costo_viaje(cant_personas, costo):
    costo_total = cant_personas*costo
    return costo_total

tipo_bus = int(input("Tipo de bus: "))
distancia = int(input("Distancia: "))


if tipo_bus == 1:
    while True:
        cant_personas = int(input("Cantidad de personas: "))
        if cant_personas > 20:
            print("\nDatos de Salida:\nCantidad no permitida\n")
            continue
        else:
            break
elif tipo_bus == 2:
    while True:
        cant_personas = int(input("Cantidad de personas: "))
        if cant_personas > 25:
            print("\nDatos de Salida:\nCantidad no permitida\n")
            continue
        else:
            break
elif tipo_bus == 3:
    while True:
        cant_personas = int(input("Cantidad de personas: "))
        if cant_personas > 30:
            print("\nDatos de Salida:\nCantidad no permitida\n")
            continue
        else:
            break


costo_persona = costoxpersona(tipo_bus, distancia)

costo_total = costo_viaje(cant_personas, costo_persona)

print(f"\nDatos de salida:\nCosto del pasaje por persona: {costo_persona}")

print(f"Costo total del viaje: {costo_total}")

```
